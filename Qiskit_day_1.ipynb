# Import libraries

```python
from qiskit import *
from qiskit.tools.monitor import job_monitor
from qiskit.tools.visualization import plot_state_qsphere
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
```

# Build simple circuit
### Explain aspects of circuit
### Classical bit
### Measurement
### Initialize a custom state

```python
# create circuit
circuit = QuantumCircuit(1, 1)

# apply gate
circuit.x(0)

# apply measurment
circuit.measure(0, 0)

# draw circuit
circuit.draw(initial_state=True, output='mpl')
```

# Run on simulator

```python
# select simulator
simulator = Aer.get_backend('aer_simulator')

# save statevector 
circuit.save_statevector()

# assemble circuit
qobj = assemble(circuit)

# Run on simulator
# job = simulator.run(circuit.decompose().decompose())
job = simulator.run(circuit)
# job =  execute(circuit, simulator)

# monitor job
job_monitor(job)

# get result
result = job.result()
```

# Statevector output

```python
# get final state
final_state = result.get_statevector(circuit)

# plot statevcector
plot_bloch_multivector(final_state)
```

# Initialize a custom state

```python
# create circuit
circuit = QuantumCircuit(1, 1)

# define a state
custom_state = '1'

# assign custom state
circuit.initialize(custom_state, 0)

# apply measurment
circuit.measure(0, 0)

# draw circuit
circuit.draw(initial_state=True, output='mpl')
```
